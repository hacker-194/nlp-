```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.txt", sep =';' , header= None , names = ['target','emotions'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["target"] = df['target'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
emotion = df['emotions'].unique()
emo = { }
i = 0
for e in emotion:
  emo[e] = i
  i += 1



```

```python

df['emotions'] = df["emotions"].map(emo)
```

```python
df
```

```python
import string

def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df["target"] = df['target'].apply(remove_punc)
```

```python
df
```

```python
df['target'] = df['target'].str.replace(r'\d+', '',regex=True)
```

```python
def remove_emojis(txt):
  new = ""
  for i in txt:
    if i.isascii():
      new += i
  return new
```

```python
df["target"] = df['target'].apply(remove_emojis)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df.loc[1]['target']
```

```python

df["target"] = df['target'].apply( lambda x:' '.join([words for words in x.split() if words not in stop_words]))
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df['target']
y = df['emotions']
x_train , x_test , y_train, y_test = train_test_split(x ,y , test_size=0.2 , random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(x_train)
X_test_bow = bow_vectorizer.transform(x_test)
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
y_pred_bow = nb_model.predict(X_test_bow)
accuracy_bow = accuracy_score(y_test, y_pred_bow)
```

```python
print(accuracy_bow)
```

```python
tdf = TfidfVectorizer()
X_train_tdf = tdf.fit_transform(x_train)
X_test_tdf = tdf.transform(x_test)
```

```python
nb_model.fit(X_train_tdf, y_train)
y_pred_tdf = nb_model.predict(X_test_tdf)
accuracy_tdf = accuracy_score(y_test, y_pred_tdf)
print(  accuracy_tdf)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_bow, y_train)
```

```python
log_pred = logistic_model.predict(X_test_bow)
```

```python
print(accuracy_score(y_test,log_pred ))
```
