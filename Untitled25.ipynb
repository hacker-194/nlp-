```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
    text = str(text).lower()
    # remove puncation
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # words tockenization
    words = text.split()
    #remove stopwords
    words = [word for word in words if word not in stop_words]
    words = [lemmatizer.lemmatize(word) for word in words]

    return " ".join(words)
```

```python
df['tag'] = df['tag'].apply(preprocess_text)
```

```python
df['tag'][0]
```

```python
 df = df.reset_index(drop = True)
```

```python
indices = pd.Series(df.index, index=df['title'])
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1,2) , stop_words='english')
```

```python
trfid_matrix = tfidf.fit_transform(df['tag'])
```

```python
trfid_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommended( title , n= 10):
  if title not in indices:
    return ['movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(trfid_matrix[idx], trfid_matrix).flatten()
  sim_index = sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[sim_index]


```

```python
recommended('The Dark Knight Rises')
```

```python
import pickle
pickle.dump(trfid_matrix, open('trfid_matrix.pkl', 'wb'))
pickle.dump(indices, open('indices.pkl', 'wb'))
df.to_pickle('df.pkl')
pickle.dump(tfidf,open('tfidf.pkl','wb'))
```
